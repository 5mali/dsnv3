```python
import numpy as np
def rewardfn(enp):
    R_PARAM = 20000 #chosen empirically for best results
    mu = 0
    sig = 0.06*R_PARAM #knee curve starts at approx. 2000mWhr of deviation
    norm_reward = 3*(np.exp(-np.power((enp - mu)/sig, 2.)/2) / np.exp(-np.power((0 - mu)/sig, 2.)/2))-1

#     if(np.abs(enp) <= 0.12*R_PARAM):
#         norm_reward = 2*(np.exp(-np.power((enp - mu)/sig, 2.)/2) / np.exp(-np.power((0 - mu)/sig, 2.)/2))
#     else:
#         norm_reward = -0.25 - 10*np.abs(enp/R_PARAM)
    return (norm_reward)
```

```python
x = np.arange(-4000,4000)
y = [rewardfn(i) for i in x]
```

```python
import matplotlib.pyplot as plt

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
DAY_SPACING = 3
NO_OF_DAYS = 20

x = np.arange(0,24*NO_OF_DAYS,1)
y = [np.abs(np.sin(i*np.pi/24)) for i in x]

tick_spacing = 24*DAY_SPACING


fig = plt.figure(figsize=(24,3))

ax = fig.add_subplot(111)
ax.plot(x,y)
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

labels = [item for item in ax.get_xticklabels()]
labels [1:-1] = np.arange(0,NO_OF_DAYS,DAY_SPACING) #the first label is reserved to negative values
ax.set_xticklabels(labels)

plt.show()
```
